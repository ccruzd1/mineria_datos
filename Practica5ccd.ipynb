### Carga y visualización del data set breast cancer 

###### Importar librerías

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import pandas as pd
```

###### Carga el data set

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

###### Dimensiones del data set

```python
print ("Cantidad de filas =", cancer.data.shape[0], "instancias", "y cantidad de columnas =", cancer.data.shape[1], "atributos.", "Las dos clases son",cancer.target_names[0],"y",cancer.target_names[1])
```

## 2. Algoritmos de clasificación

Seleccione 5 algoritmos de clasificación, procure evitar el uso de los algoritmos básicos, de hecho puede
recurrir al uso de algoritmos que no se hayan visto en clase. Para cada algoritmo construya el clasificador
usando los parámetros "estimados" o "por defecto".

## 3. Validación tradicional

Separe el conjunto de entrenamiento, del conjunto de pruebas (30% para pruebas) y determine la eficiencia
de cada método, con los valores "estimados" o "por defecto" del método.

*NOTA: para efectos de comodidad, estas dos secciones se harán en conjunto.

### Selección de 5 algoritmos de clasificación

1. Random Forest, 2. AdaBoost, 3. MLPClassifier, 4. Decision Tree y 5. Gradient Boosting

### Separación del conjunto de entrenamiento y de test

```python
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.3)
```

### Contrucción de los Clasificadores con parámetros por defecto y estimados

### 1. Random Forest

###### Parámetros por defecto:

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier().fit(X_train, y_train)
print("Training score: {:.2f}".format(rfc.score(X_train, y_train)))
print("Test score: {:.2f}".format(rfc.score(X_test, y_test)))
```

```python
%matplotlib inline

training_acc = []
test_acc = []

rango = range(1, 11)
for n_estimators in rango:
    rfc = RandomForestClassifier(n_estimators=n_estimators).fit(X_train, y_train)
    training_acc.append(rfc.score(X_train, y_train))
    test_acc.append(rfc.score(X_test, y_test))
plt.plot(rango, training_acc, label="training accuracy")
plt.plot(rango, test_acc, label="test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_estimators")
plt.legend()
```

###### Parámetros estimados:

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=4, min_samples_split=8).fit(X_train, y_train)
print("Training score: {:.2f}".format(rfc.score(X_train, y_train)))
print("Test score: {:.2f}".format(rfc.score(X_test, y_test)))
```

```python
%matplotlib inline

training_acc = []
test_acc = []

rango = range(1, 11)
for n_estimators in rango:
    rfc = RandomForestClassifier(n_estimators=n_estimators).fit(X_train, y_train)
    training_acc.append(rfc.score(X_train, y_train))
    test_acc.append(rfc.score(X_test, y_test))
plt.plot(rango, training_acc, label="training accuracy")
plt.plot(rango, test_acc, label="test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_estimators")
plt.legend()
```

con un valor de alrededor de 3 n estimators está más alejado del overfitting

###### Parámetros estimados:

```python
from sklearn.ensemble import AdaBoostClassifier
adaboost = AdaBoostClassifier(n_estimators=6).fit(X_train, y_train)
print("Training set score: {:.2f}".format(adaboost.score(X_train, y_train)))
print("Test set score: {:.2f}".format(adaboost.score(X_test, y_test)))
```

con un valor de alrededor de 4 n estimators está más alejado del overfitting

### 3. MLPClassifier

###### Parámetros por defecto:

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier().fit(X_train, y_train)
print("Training set score: {:.2f}".format(mlp.score(X_train, y_train)))
print("Test set score: {:.2f}".format(mlp.score(X_test, y_test)))
```

```python
%matplotlib inline

training_acc = []
test_acc = []

rango = range(2,8)
for hidden_layer_sizes in rango:
    mlp = MLPClassifier(hidden_layer_sizes=hidden_layer_sizes).fit(X_train, y_train)
    training_acc.append(mlp.score(X_train, y_train))
    test_acc.append(mlp.score(X_test, y_test))
plt.plot(rango, training_acc, label="training accuracy")
plt.plot(rango, test_acc, label="test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("hidden_layer_sizes")
plt.legend()
```

###### Parámetros estimados:

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(hidden_layer_sizes=(2, ), alpha=0.00001,solver='lbfgs').fit(X_train, y_train)
print("Training set score: {:.2f}".format(mlp.score(X_train, y_train)))
print("Test set score: {:.2f}".format(mlp.score(X_test, y_test)))
```

```python
%matplotlib inline

training_acc = []
test_acc = []

rango = range(2,8)
for hidden_layer_sizes in rango:
    mlp = MLPClassifier(hidden_layer_sizes=hidden_layer_sizes,solver='lbfgs').fit(X_train, y_train)
for alpha in rango:
    mlp = MLPClassifier(alpha=alpha).fit(X_train, y_train)
    training_acc.append(mlp.score(X_train, y_train))
    test_acc.append(mlp.score(X_test, y_test))
plt.plot(rango, training_acc, label="training accuracy")
plt.plot(rango, test_acc, label="test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("hidden_layer_sizes")
plt.legend()
```

con un valor de alrededor de 2 hidden layers está más alejado del overfitting

En particular este algoritmo presenta menos consistencia

## 4. Validación Cruzada

Utilice dos de los métodos de Cross-Validation y aplíquelos a cada método de clasificación

###### Shuffle-Split CV

```python
from sklearn.model_selection import cross_val_score, ShuffleSplit

shuffle_split = ShuffleSplit(test_size=.3, train_size=.7, n_splits=10)

rfc = RandomForestClassifier(n_estimators=4, min_samples_split=8)
adaboost = AdaBoostClassifier(n_estimators=6)
mlp = MLPClassifier(hidden_layer_sizes=(250, ), alpha=0.00001,solver='lbfgs')
dtc = DecisionTreeClassifier(max_depth=2)
gbc = GradientBoostingClassifier(max_depth=2)

algoritmos=[rfc, adaboost, mlp, dtc, gbc]

for i in algoritmos:
    scores = cross_val_score(i, cancer.data, cancer.target, cv=shuffle_split)
    print("Cross-validation scores de",i,": {}".format(scores))
    print("Average cross-validation score: {:.2f}".format(scores.mean()))
```

###### Cross Validation (K-fold)

```python
from sklearn.model_selection import cross_val_score

rfc = RandomForestClassifier(n_estimators=3)
adaboost = AdaBoostClassifier(n_estimators=6)
mlp = MLPClassifier(hidden_layer_sizes=(250, ), alpha=0.00001,solver='lbfgs')
dtc = DecisionTreeClassifier(max_depth=2)
gbc = GradientBoostingClassifier(max_depth=1)

algoritmos=[rfc, adaboost, mlp, dtc, gbc]

for i in algoritmos:
    scores = cross_val_score(i, cancer.data, cancer.target)
    print("Cross-validation scores de",i,": {}".format(scores))
    print("Average cross-validation score: {:.2f}".format(scores.mean()))
```

## 5. Conclusiones

Utilizando los algoritmos de clasificación con parámetros por defecto demuestra en general la necesidad de optimizar parámetros para obtener mejores resultados. Específicamente Random Forest, AdaBoost, Decision Tree y Gradient Boosting tendieron al overfitting, mientras que el rendimiento de MLPClassifier de forma opuesta, es muy bajo en cuanto a precisión.

A partir de la optimización de parámetros, mejoró la precisión comparable entre el set de entrenamiento y el de prueba con todos los algoritmos, sin embargo, MLPClassifier fue el algoritmo con menor estabilidad demostrada visualmente en el gráfico.

Las herramientas de validación cruzada demostraron valores aceptables que son comparables a través de ambos métodos utilizando los mismos parámetros asignados a los algoritmos de clasificación utilizados.
